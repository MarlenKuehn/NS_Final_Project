```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the GraphML file
G = nx.read_graphml('../dataset/morphological_parrot_network.graphml')

# Visualize the graph
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```
